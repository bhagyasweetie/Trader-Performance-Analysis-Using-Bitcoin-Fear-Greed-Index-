```python
import pandas as pd

fear = pd.read_csv("fear_greed_index (1) (1).csv")
trades = pd.read_csv("historical_data.csv")

print("Fear & Greed Shape:", fear.shape)
print("Trades Shape:", trades.shape)
```

```python
print("Fear Dataset")
display(fear.head())

print("\nColumns:")
print(fear.columns)

print("\nTrader Dataset")
display(trades.head())

print("\nColumns:")
print(trades.columns)
```

```python
print("Fear Dataset Columns:")
print(fear.columns)

display(fear.head())
```

```python
trades['Timestamp IST'] = pd.to_datetime(
    trades['Timestamp IST'],
    dayfirst=True
)
```

```python
fear['date'] = pd.to_datetime(fear['date'])
```

```python
print(trades['Timestamp IST'].head())
```

```python
fear['Date'] = fear['date'].dt.date
trades['Date'] = trades['Timestamp IST'].dt.date
```

```python
merged = pd.merge(
    trades,
    fear[['Date', 'classification', 'value']],
    on='Date',
    how='left'
)
```

```python
merged[['Date', 'Closed PnL', 'classification', 'value']].head()
```

```python
merged.groupby('classification')['Closed PnL'].mean().sort_values(ascending=False)
```

```python
merged.groupby('classification')['Closed PnL'].sum().sort_values(ascending=False)
```

```python
merged['Win'] = merged['Closed PnL'] > 0

merged.groupby('classification')['Win'].mean() * 100
```

```python
merged['classification'].value_counts()
```

```python
merged.groupby('classification')['Closed PnL'].sum().sort_values(ascending=False)
```

```python
merged['Win'] = merged['Closed PnL'] > 0

merged.groupby('classification')['Win'].mean() * 100
```

```python
import matplotlib.pyplot as plt

merged['classification'].value_counts().plot(kind='bar')

plt.title('Number of Trades by Sentiment')
plt.ylabel('Trade Count')
plt.show()
```

```python
win_rate = merged.groupby('classification')['Win'].mean()*100

win_rate.sort_values().plot(kind='bar')

plt.title('Win Rate by Sentiment')
plt.ylabel('Win Rate (%)')
plt.show()
```

```python
merged.groupby('classification')['Closed PnL'].mean().sort_values().plot(kind='bar')

plt.title('Average Closed PnL by Sentiment')
plt.ylabel('Average Profit')
plt.show()
```
